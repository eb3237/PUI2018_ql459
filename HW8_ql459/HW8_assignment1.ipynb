### I did the analysis on the distribution of trip duration of yellow taxi using the 2017 Yellow Taxi Trip Data

```python
url = "https://data.cityofnewyork.us/resource/su3p-fcfz.csv"
yellowtaxi = pd.read_csv(url)
```

```python
yellowtaxi.groupby('tpep_pickup_datetime')
yellowtaxi.head()
```

```python
yellowtaxi.columns
```

```python
yellowtaxi.dropna(inplace=True)

yellowtaxi['tpep_pickup_datetime'] = pd.to_datetime(yellowtaxi['tpep_pickup_datetime'])
yellowtaxi['tpep_dropoff_datetime'] = pd.to_datetime(yellowtaxi['tpep_dropoff_datetime'])

yellowtaxi['trip_duration'] = yellowtaxi['tpep_dropoff_datetime'] - yellowtaxi['tpep_pickup_datetime']

yellowtaxi['trip_duration'] = pd.to_timedelta(yellowtaxi['trip_duration'])
```

```python
yellowtaxi.head()
```

```python
# yellowtaxi['trip_duration_second'] = (yellowtaxi['trip_duration'].dt.second.div(60).astype(int) 
           #  + yellowtaxi['trip_duration'].dt.day.multiply(1440).astype(int) )


#yellowtaxi['trip_duration'] = pd.to_timedelta(yellowtaxi['trip_duration'])
#yellowtaxi['trip_duration_second'] = (yellowtaxi['trip_duration'].dt.second().astype(int)
#                                      + yellowtaxi['trip_duration'].dt.minutes().multiply(60).astype(int))
yellowtaxi['trip_duration_second'] = yellowtaxi['trip_duration'].dt.total_seconds().astype(int)
```

```python
yellowtaxi.head()
```

```python
fig, ax = plt.subplots(figsize=(12,10))
ax = sns.distplot(yellowtaxi.trip_duration_second,bins=60,kde=True,hist=True,color='r', label='trip duration')
plt.title("The trip duration distribution of NYC green taxi on 10am in Nov. 15th, 2017", fontsize=20)
plt.xlabel('Trip duration (in second)', fontsize = 16)
plt.ylabel('Frequency', fontsize = 16)
plt.legend()
plt.show
```
