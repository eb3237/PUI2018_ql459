Analysis how the trip duration is changing with time for different day in the week

```python
taxi.dropna(inplace=True)

taxi['tpep_dropoff_datetime'] = pd.to_datetime(taxi['tpep_dropoff_datetime'])
taxi['tpep_pickup_datetime'] = pd.to_datetime(taxi['tpep_pickup_datetime'])
taxi['trip_duration'] = taxi['tpep_dropoff_datetime'] - taxi['tpep_pickup_datetime']
taxi['trip_duration'] = pd.to_datetime(taxi['trip_duration'])
```

```python
taxi['pickup_day'] = taxi.tpep_pickup_datetime.dt.weekday_name
taxi['pickup_hour'] = taxi.tpep_pickup_datetime.dt.hour
taxi.groupby('tpep_pickup_datetime')
taxi
```
